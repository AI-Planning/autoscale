```python

DOMAINS = sorted([
    #"barman", "blocksworld", "childsnack", "data-network", "depots", "driverlog",
    #"elevators", "floortile", "grid", "gripper", "hiking", "logistics",
    #"miconic", "nomystery", "openstacks", "parking",  "satellite",
    #"scanalyzer", "snake", "storage", "tpp", "transport", "visitall",
    # "zenotravel",  "woodworking",
    "driverlog", "rovers", "snake", "grid"
])


def get_cov(name, planners, domains):
    coverage = [0 for p in planners]
    coverage_ipc = [0 for p in planners]
    
    
    for domain in domains: 
        for i, planner in enumerate(planners): 
            coverage_ipc[i] += list(df[df['track'] == "sat-1210"][df['domain'] == domain]["coverage-ipc"])[0][planner]
            coverage[i] += list(df[df['track'] == "sat-1210"][df['domain'] == domain]["coverage"])[0][planner]
    
    return ([name] + coverage_ipc + coverage)

planners = [#"fd1906-gbfs-ff",
    "fd1906-lama-first",
    #"ipc2014-agl-mpc",
    #"ipc2014-agl-mercury",
    #"ipc2014-agl-jasper",
    #"ipc2014-agl-probe",
    #"ipc2018-agl-decstar",
   # "ipc2018-agl-fd-remix",
   "ipc2018-agl-lapkt-dual-bfws",
   # "ipc2018-agl-lapkt-bfws-pref",
   # "ipc2018-agl-lapkt-poly-bfws",
    "ipc2018-agl-olcff",
    #"ipc2018-agl-saarplan",
   # "ipc2018-agl-cerberus",
]


df_coverage_table = pd.DataFrame([get_cov (domain, planners, [domain]) for domain in DOMAINS] + [get_cov ("Total", planners, DOMAINS)] , columns=["domain"] + [f"IPC {p}" for p in planners]+ [p for p in planners])
print(df_coverage_table.to_latex())
```
