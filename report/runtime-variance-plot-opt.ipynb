```python
import pandas as pd
```

```python
df = pd.read_json("../optimization/results/2020-11-25-A-runtime-variance-opt-properties.json").transpose()
df.head()
```

```python
df2 = df[["algorithm", "domain", "problem", "runtime"]]
df2["runtime"] = df2["runtime"].astype(float)
df2
```

```python
df2['problemgroup'] = df2["problem"].str.extract(r'(?P<problemgroup>p\d\d)-\d\.pddl', expand=False)
df2["seedgroup"] = df2["problem"].str.extract(r'p\d\d-(?P<seedgroup>\d)\.pddl', expand=False)
df2 = df2.dropna()

#with open("all-runtimes.txt", "w") as f:
#    f.write(df2.to_string())
df2
```

```python
df3 = df2.sort_values(["algorithm", "domain", "seedgroup", "runtime"])

s_index_values = []
for gname, gdata in df3.groupby(["algorithm", "domain", "seedgroup"]):
    s_index_values += range(len(gdata))
df3['sindex'] = s_index_values

df3
```

```python

for gname, gdata in df3.groupby(["domain", "algorithm"]):
    
    dfplotpivot = gdata.pivot(index='sindex', columns='seedgroup', values='runtime')

    #ax = dfplotpivot.plot(title=str(gname), logy=True)
```

```python
df4 = df2.groupby(["domain", "problem", "seedgroup"])["runtime"].aggregate(["min"])

df4 = df4.dropna()

df4 = df4.reset_index()
df4 = df4.sort_values(["domain", "seedgroup", "min"])


s_index_values = []
for gname, gdata in df4.groupby(["domain", "seedgroup"]):
    s_index_values += range(len(gdata))
df4['sindex'] = s_index_values


for gname, gdata in df4.groupby(["domain"]):
    dfplotpivot = gdata.pivot(index='sindex', columns='seedgroup', values='min')
    dfplotpivot.plot(title=str(gname), logy=True)


#df4.head(100)
```
