```python
import pandas as pd
```

```python
df = pd.read_json("../optimization/results/2020-11-25-A-runtime-variance-opt-properties.json").transpose()
df.head()
```

```python
df2 = df[["algorithm", "domain", "problem", "runtime"]]
df2["runtime"] = df2["runtime"].astype(float)
df2
```

```python
pg = df2["problem"].str.extract(r'(?P<problemgroup>p\d\d)-\d\.pddl', expand=False)
df2["problemgroup"] = pg
df2 = df2.dropna()

with open("all-runtimes.txt", "w") as f:
    f.write(df2.to_string())
df2
```

```python
df3 = df2.groupby(["algorithm", "domain", "problemgroup"])["runtime"].aggregate(["count", "mean", "std", "min", "max"])
print(len(df3))
df3.plot(x ='mean', y='max', kind = 'scatter',logx=True,logy=True)
df3.plot(x ='mean', y='min', kind = 'scatter',logx=True,logy=True)
df3.plot(x ='max', y='min', kind = 'scatter',logx=True,logy=True)

df3 = df3[lambda x : x["max"] - x["min"] > 50]
print(len(df3))
df3 = df3[lambda x : x["max"] -x["min"] > 100]
print(len(df3))
with open("runtime-variance-opt.txt", "w") as f:
    f.write(df3.to_string())
df3


```
